```python
# 1. Importo el modulo de pandas con el alias pd
import pandas as pd

# 2. Cargo el archivo de Excel en una variable de nombre: archivo
archivo = pd.ExcelFile("DATOS_2025_SEM_2.xlsx")
```

```python
archivo.sheet_names
```

```python
housings = archivo.parse("Housing")
housings
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 7))
plt.scatter(housings['longitude'], housings['latitude'], alpha=0.5)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Geographic Distribution of Housing Data')
plt.grid(True)
plt.show()
```
